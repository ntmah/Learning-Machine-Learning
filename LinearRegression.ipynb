```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area, df.price, color = 'red', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python

f = pd.read_csv('areas.csv')
```

```python
# Du doan gia nha trong tep areas.csv
p = reg.predict(f)
```

```python
f['prices'] = p
f
```

```python
f.to_csv('prediction.csv')
```
